```python
import numpy as np
import tensorflow as tf
from urllib import request
```

```python
example3_url="https://raw.githubusercontent.com/CSC-CONICET/CURSO-LEARNING1/main/datasets/glass.csv"
f = request.urlopen(example3_url)
dataset = np.genfromtxt(f, delimiter=',',skip_header=1)
```

```python
ntr = dataset.shape[0]
dx = dataset.shape[1]-1
original_classes = np.unique(dataset[:,-1])
num_classes = original_classes.shape[0]
labels = np.zeros_like(dataset[:,-1])
for k in range(num_classes):
  labels[dataset[:,-1]==original_classes[k]]=k
  
mu = np.mean(dataset[:,:-1],axis=0)
sigma = np.std(dataset[:,:-1],axis=0)
dataset0=(dataset[:,:-1]-mu)/sigma
```

```python
input = tf.keras.layers.Input(shape=(dx,), name='Entrada')
x1 = tf.keras.layers.Dense(num_classes-1, name='wxmasb')(input)
out = tf.keras.layers.Concatenate(name='OpConc')([x1, tf.zeros_like(input[:,:1])])
model = tf.keras.models.Model(inputs=input, outputs=out)
model.summary()
```

```python
es = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=1)
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=3.5),loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
hist = model.fit(x=dataset0, y=labels, batch_size = ntr, epochs=1000,callbacks=[es])
```

```python
def reinitialize(model):
    for l in model.layers:
        if hasattr(l,"kernel_initializer"):
            l.kernel.assign(l.kernel_initializer(tf.shape(l.kernel)))
        if hasattr(l,"bias_initializer"):
            l.bias.assign(l.bias_initializer(tf.shape(l.bias)))
        if hasattr(l,"recurrent_initializer"):
            l.recurrent_kernel.assign(l.recurrent_initializer(tf.shape(l.recurrent_kernel)))
```

```python
param = model.get_weights()
print(param)
```

```python
reinitialize(model)
param = model.get_weights()
print(param)
```
